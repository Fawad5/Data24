### ***`Import` `Libraries`***

```python
# To download datasets from online sources like (Kaggle, Google Drive) directly into your Colab environment
import opendatasets as od

# To handle data
import pandas as pd
import numpy as np

# To visualize data
import matplotlib.pyplot as plt
import seaborn as sns

# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# To download datasets from Kaggle

od.download("https://www.kaggle.com/datasets/uciml/red-wine-quality-cortez-et-al-2009")
```

```python
# Load dataset

df = pd.read_csv("red-wine-quality-cortez-et-al-2009/winequality-red.csv")
df.head()
```

```python
# Shape of dataset

df.shape
```

```python
print(f'Number of rows: {df.shape[0]}')
print(f'Number of columns: {df.shape[1]}')
```

```python
# Columns' name in the dataset

df.columns
```

```python
# Information about dataset

df.info()
```

```python
# Summary

df.describe()
```

```python
# Check Null Values

df.isnull().sum()
```

```python
print('So we have no null values in the dataset.')
```

```python
sns.histplot(df['quality'])
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
# Check duplicate vlues

df.duplicated().sum()
```

```python
print(f'So we have {df.duplicated().sum()} values in the dataset.')
```

```python
# Remove duplicate rows

df.drop_duplicates(inplace=True)
```

```python
# Check duplicate vlues

df.duplicated().sum()
```

```python
df.head()
```

```python
quality_mean  = df.groupby('quality').mean()
quality_mean
```

```python
# chart

plt.figure(figsize=(18, 6))
sns.barplot(quality_mean, palette='Set2',ci=None)
plt.title('Quality of Wine based on physicochemical properties',fontsize=12, fontweight='bold',fontstyle='italic')
plt.xlabel('Quality', fontsize=10, fontstyle='italic')
plt.ylabel('Count', fontsize=10, fontstyle='italic')
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
sns.despine()
plt.grid(axis='y', linestyle=':', alpha=0.4)
plt.show()
```

```python
sns.histplot(quality_mean, color='red')
```

```python
quality_mean.corr()
```

```python
plt.figure(figsize=(20, 7))
sns.heatmap(df.corr(), annot=True, cmap='Blues',fmt='f')
```

***In this Heatmap `alcohol` have a `positive` correlation with quality, indicating that higher alcohol content tends to be associated with better wine quality.***

---


***And `volatile acidity` have a `negative` correlation with quality, indicating that higher volatile acidity is associated with lower wine quality.***

```python
plt.figure(figsize=(20, 7))
sns.heatmap(quality_mean.corr(), annot=True, cmap='Blues',fmt='f')
plt.title("Heatmap (Wines' Quality based on physicochemical properties)",fontsize=14, fontweight='bold',fontstyle='italic')

plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
sns.despine()
plt.show()

```

***This Heatmap shows `Wine's Quality based on physiochamical properties`. Wine Quality is `positively` correlated with `sulphates`, `citric acid`, represents that higher content of these features affect wine's quality.***
